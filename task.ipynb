```python
# Identify overall performance of the store based on year wise
```

```python
import pandas as pd
```

```python
path2='/content/sales_data_set.csv'
df2=pd.read_csv(path2)
df2

```

```python
path3='/content/Features_data_set.csv'
df3=pd.read_csv(path3)
df3

```

```python
df=pd.concat([df2,df3],axis=0)
df['year'] = pd.DatetimeIndex(df['Date']).year
stores=[]
for i in range(1,46):
  print("store number is",i)
  bb1=df3[df3['Store']==i]
  Year_cpi= df.groupby('year')['CPI'].sum()
  Yearly_cpi=Year_cpi.reset_index()
  Year_sale=df.groupby('year')['Weekly_Sales'].sum()
  Yearly_sale=Year_sale.reset_index()
  store=pd.concat([Yearly_sale,Yearly_cpi],axis=1)
  store['est_sale']=store['Weekly_Sales']+store['CPI']
  store['performance']=store['Weekly_Sales']/store['est_sale']
  print(store)
  print("-------------------------------------------*****************-------------------------------------------")
```
